```python
import numpy as np

try:
    # Print the sum of two None's
    print("Add operation output of 'None': ", None + None)

except TypeError:
    # Print if error
    print("'None' does not support Arithmetic Operations!!")

try:
    # Print the sum of two np.nan's
    print("Add operation output of 'np.nan': ", np.nan + np.nan)

except TypeError:
    # Print if error
    print("'np.nan' does not support Arithmetic Operations!!")

try:
    # Print the output of logical OR of two None's
    print("OR operation output of 'None': ", None or None)

except TypeError:
    # Print if error
    print("'None' does not support Logical Operations!!")

try:
    # Print the output of logical OR of two np.nan's
    print("OR operation output of 'np.nan': ", np.nan or np.nan)

except TypeError:
    # Print if error
    print("'np.nan' does not support Logical Operations!!")
```

`None` is of type `NoneType` while `np.nan` is of `float`. This allows `np.nan` to have both arithmetic and logical operations.

# **Missingness across a variable**

* `matplotlib` skips all missing values while plotting
* We would need to first create a function that fills in dummy values for all the missing values in the DataFrame before plotting
* The `numpy.random.rand` generates unifrom(0, 1) so we need to multiply them by the range of variables of interest

### **Nullity**

* sum of null values of both respective columns that you are plotting calculated using the `.isnull()` method
* The `nullity` returns a Series of `True` or `False` where:

  * `True` -->  Either `col1` or `col2` or both values are missing.
  * `False`-->  Neither of `col1` and `col2` values are missing.

### จากกราฟ สีแดงคือ ค่าข้อมูลที่หาย

```python
import matplotlib.pyplot as plt
import pandas as pd

def fill_dummy_values(df, scaling_factor=0.075):
    df_dummy = df.copy(deep=True)
    for col_name in df_dummy:
        col = df_dummy[col_name]
        col_null = col.isnull()    
        # Calculate number of missing values in column 
        num_nulls = col_null.sum()
        # Calculate column range
        col_range = col.max() - col.min()
        # Scale the random values to scaling_factor times col_range
        dummy_values = (rand(num_nulls) - 2) * scaling_factor * col_range + col.min()
        col[col_null] = dummy_values
    return df_dummy


diabetes = pd.read_csv("pima-indians-diabetes.csv")

# matplotlib skips all missing values while plotting
diabetes.plot(kind="scatter", x="Serum_Insulin", y="BMI")
plt.show()

# fills in dummy values
from numpy.random import rand

# Create dummie
dummy_values = rand(diabetes["BMI"].isnull().sum())
print(dummy_values)

# Shift dummies to (-2, -1)
dummy_values = dummy_values - 2

# Scale to 0.075 (This number is from trial and error) of Column Range
BMI_range = diabetes["BMI"].max() - diabetes["BMI"].min()

# Shift to column minimum
dummy_values = dummy_values * 0.075 * BMI_range + diabetes["BMI"].min()
print("\n", dummy_values)

diabetes_dummy = fill_dummy_values(diabetes)

# Get missing values of both columns for coloring
# the sum of null values of both respective columns that you are plotting
nullity = diabetes["Serum_Insulin"].isnull() + diabetes["BMI"].isnull()
fig, ax = plt.subplots()
diabetes_dummy.plot(ax=ax, kind="scatter", alpha=0.5, c=nullity, cmap="rainbow", x="Serum_Insulin", y="BMI")
plt.show()


# Fill dummy values in diabetes_dummy
diabetes_dummy = fill_dummy_values(diabetes)

# Sum the nullity of Skin_Fold and BMI
nullity = diabetes["Skin_Fold"].isnull() + diabetes["BMI"].isnull()

# Create a scatter plot of Skin Fold and BMI 
fig, ax = plt.subplots()
diabetes_dummy.plot(ax=ax, x="Skin_Fold", y="BMI", kind='scatter', alpha=0.5, c=nullity, cmap='rainbow')

plt.show()
```

# **When to delete missing values**

## **Can delete only when data are missing COMPLETELY AT RANDOM**

1.   **Pairwise deletion**: can skip missing in aggregation คือการหา aggregation มันก็ skip ให้อยู่แล้ว

2.   **Listwise deletion**: delete entire rows
  * `df.dropna(subset=["col_name_to_delete"], how="any", inplace=True)`



```python
import pandas as pd
import missingno as msno

diabetes = pd.read_csv("pima-indians-diabetes.csv")

# Delete every missing values results in much smaller data
display(msno.matrix(diabetes.dropna()))

# Drop only Glucose 
diabetes = pd.read_csv("pima-indians-diabetes.csv")
diabetes.dropna(subset=["Glucose"], how="any", inplace=True)

# Visualize the missingness of diabetes after dropping missing values
display(msno.matrix(diabetes))

# Drop only Glucose with ALL
diabetes = pd.read_csv("pima-indians-diabetes.csv")
diabetes.dropna(subset=["Glucose"], how="all", inplace=True)

# Visualize the missingness of diabetes after dropping missing values
display(msno.matrix(diabetes))
```

# **Mean, Median, Mode, Constant Imputation**

* Not recommended as it makes data bias
* The imputed values lie as a straigtline on y axis, this means that they are not adjusted to the correlation of the data



```python
from sklearn.impute import SimpleImputer
import pandas as pd
import matplotlib.pyplot as plt

diabetes = pd.read_csv("pima-indians-diabetes.csv")

# Mean
diabetes_mean = diabetes.copy(deep=True)
mean_imputer = SimpleImputer(strategy="mean")
diabetes_mean.iloc[:, :] = mean_imputer.fit_transform(diabetes_mean)

# Median
diabetes_median = diabetes.copy(deep=True)
median_imputer = SimpleImputer(strategy="median")
diabetes_median.iloc[:, :] = median_imputer.fit_transform(diabetes_median)

# Mode
diabetes_mode = diabetes.copy(deep=True)
mode_imputer = SimpleImputer(strategy="most_frequent")
diabetes_mode.iloc[:, :] = mode_imputer.fit_transform(diabetes_mode)

# Constant
diabetes_constant = diabetes.copy(deep=True)
constant_imputer = SimpleImputer(strategy="constant", fill_value=0)
diabetes_constant.iloc[:, :] = constant_imputer.fit_transform(diabetes_constant)

# Scatter Plot
nullity = diabetes["Serum_Insulin"].isnull() + diabetes["Glucose"].isnull()
fig, ax = plt.subplots()
diabetes_mean.plot(ax=ax, x='Serum_Insulin', y="Glucose", kind="scatter", alpha=0.5, c=nullity, cmap="rainbow", title="Mean Imputation", grid=True)
plt.show()

# Set nrows and ncols to 2
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))
nullity = diabetes.Serum_Insulin.isnull()+diabetes.Glucose.isnull()

# Create a dictionary of imputations
imputations = {'Mean Imputation': diabetes_mean, 'Median Imputation': diabetes_median, 
               'Most Frequent Imputation': diabetes_mode, 'Constant Imputation': diabetes_constant}

# Loop over flattened axes and imputations
for ax, df_key in zip(axes.flatten(), imputations):
    # Select and also set the title for a DataFrame
    imputations[df_key].plot(x='Serum_Insulin', y='Glucose', kind='scatter', 
                          alpha=0.5, c=nullity, cmap='rainbow', ax=ax, 
                          colorbar=False, title=df_key)
plt.show()
```

# **Imputing Categorical Data**

1.   Convert non-missing categorical columns to ordinal values
2.   Impute the missing values in the ordinal DataFrame
3. Convert Back from ordinal values to categorical values



```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder

users = pd.read_csv("users.csv").drop("Unnamed: 0", axis=1)
users.columns = ["smoker", "drink_level", "dress_preference",  "ambience", "hijos", "activity",  "budget"]

display(users.head(50))

# See the null values
ambience_null = users["ambience"].isnull()
print(users["ambience"][ambience_null])

# Create Ordinal encoder
ambience_ord_enc = OrdinalEncoder()

# Select non-null values of ambience column in users
ambience = users['ambience']
ambience_not_null = ambience[ambience.notnull()]

# Reshape ambience_not_null to shape (-1, 1)
reshaped_vals = ambience_not_null.values.reshape(-1, 1)

# Ordinally encode reshaped_vals
encoded_vals = ambience_ord_enc.fit_transform(reshaped_vals)

# Assign back encoded values to non-null values of ambience in users
users.loc[ambience.notnull(), 'ambience'] = np.squeeze(encoded_vals)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder

users = pd.read_csv("users.csv").drop("Unnamed: 0", axis=1)
users.columns = ["smoker", "drink_level", "dress_preference",  "ambience", "hijos", "activity",  "budget"]

# Create an empty dictionary ordinal_enc_dict
ordinal_enc_dict = {}

for col_name in users:
    # Create Ordinal encoder for col
    ordinal_enc_dict[col_name] = OrdinalEncoder()
    col = users[col_name]
    
    # Select non-null values of col
    col_not_null = col[col.notnull()]
    reshaped_vals = col_not_null.values.reshape(-1, 1)
    encoded_vals = ordinal_enc_dict[col_name].fit_transform(reshaped_vals)
    
    # Store the values to non-null values of the column in users
    users.loc[col.notnull(), col_name] = np.squeeze(encoded_vals)

users
```

# **KNN Categorical Imputer**



```python
from fancyimpute import KNN
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder

# users must be ordinally encoded from the previous steps

# Create KNN imputer
KNN_imputer = KNN()

# Impute and round the users DataFrame
users.iloc[:, :] = np.round(KNN_imputer.fit_transform(users))

# Loop over the column names in users
for col_name in users:
    
    # Reshape the data
    reshaped = users[col_name].values.reshape(-1, 1)
    
    # Perform inverse transform of the ordinally encoded columns
    users[col_name] = ordinal_enc_dict[col_name].inverse_transform(reshaped)

users.head(50)
```

# **Use linear regression for evaluating imputations**

```python
import statsmodels.api as sm
from sklearn.impute import SimpleImputer
from fancyimpute import KNN, IterativeImputer
import pandas as pd
import matplotlib.pyplot as plt

diabetes = pd.read_csv("pima-indians-diabetes.csv")

# Mean
diabetes_mean = diabetes.copy(deep=True)
mean_imputer = SimpleImputer(strategy="mean")
diabetes_mean.iloc[:, :] = mean_imputer.fit_transform(diabetes_mean)

# knn

knn_imputer = KNN()
diabetes_knn = diabetes.copy(deep=True)
diabetes_knn.iloc[:,:] = knn_imputer.fit_transform(diabetes_knn)

# mice

MICE_imputer = IterativeImputer()
diabetes_MICE = diabetes.copy(deep=True)
diabetes_MICE.iloc[:, :] = MICE_imputer.fit_transform(diabetes_MICE)

# Model where we drop all missing values
diabetes_cc = diabetes.dropna(how="any")
X = sm.add_constant(diabetes_cc.iloc[:, :-1])
y = diabetes_cc["Class"]
lm = sm.OLS(y, X).fit()

# Mean Imputation regression

X = sm.add_constant(diabetes_mean.iloc[:,:-1])
y = diabetes["Class"]
lm_mean = sm.OLS(y, X).fit()

# KNN Imputation
X = sm.add_constant(diabetes_knn.iloc[:,:-1])
lm_KNN = sm.OLS(y, X).fit()

# MICE Imputation
X = sm.add_constant(diabetes_MICE.iloc[:,:-1])
lm_MICE = sm.OLS(y, X).fit()

print([lm.rsquared_adj, lm_mean.rsquared_adj, lm_KNN.rsquared_adj, lm_MICE.rsquared_adj])

# Check the distribution

diabetes_cc["Skin_Fold"].plot(kind="kde", c="red", linewidth=3)
diabetes_mean["Skin_Fold"].plot(kind="kde")
diabetes_knn["Skin_Fold"].plot(kind="kde")
diabetes_MICE["Skin_Fold"].plot(kind="kde")

labels = ["Baseline (Complete Case)", "Mean Impute", "KNN Impute", "MICE Impute"]
plt.legend(labels)
plt.xlabel("Skin Fold")
plt.show()
```
